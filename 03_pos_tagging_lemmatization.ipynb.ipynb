```python
import re
import glob
```

```python
#regex distance et vitesse
REG_DIST = r"\b\d+(?:[\.,]\d+)?\s*(?:km|kilomètre|kilomètres|m|mètre|mètres)\b"
REG_SPEED = r"\b\d+(?:[\.,]\d+)?\s*(?:km/h|kmh|m/s|m·s-1)\b"

def extract_dist_speed(text):
    distances = re.findall(REG_DIST, text, flags=re.IGNORECASE)
    speeds = re.findall(REG_SPEED, text, flags=re.IGNORECASE)
    return distances, speeds
```

```python
#lit lesfichiers du corpus
txt_files = sorted(glob.glob("CorpusARIA/TXT/*.txt"))

all_distances = []
all_speeds = []

for path in txt_files:
    with open(path, "r", encoding="utf-8") as f:
        text = f.read()       
    dist, speed = extract_dist_speed(text)
    all_distances += dist
    all_speeds += speed
```

```python
print("Nombre total de distances détectées :", len(all_distances))
print("Nombre total de vitesses détectées  :", len(all_speeds))

print("\n=== DÉTAIL DES DISTANCES ===")
for d in all_distances:
    print("DISTANCE :", d)

print("\n=== DÉTAIL DES VITESSES ===")
for v in all_speeds:
    print("VITESSE :", v)
```
